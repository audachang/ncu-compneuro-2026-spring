```python
from nilearn import datasets, decoding, plotting
from sklearn.svm import SVC
from sklearn.model_selection import KFold
import pandas as pd
```

## 1. Data Loading (Haxby Dataset)

```python
haxby_dataset = datasets.fetch_haxby()
func_filename = haxby_dataset.func[0]
mask_filename = haxby_dataset.mask
```

## 2. Decoding

```python
decoder = decoding.Decoder(
    estimator=SVC(kernel='linear'),
    mask=mask_filename,
    standardize=True
)

# Need behavior targets/labels to fit
# decoder.fit(func_filename, targets)
# prediction = decoder.predict(func_filename)
```
