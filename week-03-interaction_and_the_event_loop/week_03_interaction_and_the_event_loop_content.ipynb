```python
from psychopy import visual, core, event, data
import pandas as pd
import numpy as np
```

## 1. Trial Structure

```python
trials = [
    {'text': 'RED', 'color': 'red', 'congruent': 1},
    {'text': 'GREEN', 'color': 'red', 'congruent': 0},
    {'text': 'BLUE', 'color': 'blue', 'congruent': 1}
]
```

## 2. The Loop

```python
# Pseudo-code for running locally. 
# Would require a visual.Window to be open.

# win = visual.Window(...)
results = []

for trial in trials:
    # stim = visual.TextStim(win, text=trial['text'], color=trial['color'])
    # stim.draw()
    # win.flip()
    # t0 = core.getTime()
    # keys = event.waitKeys(keyList=['r', 'g', 'b', 'q'], timeStamped=True)
    
    # Mock data for notebook execution
    keys = [['r', 0.5]] 
    
    resp, rt = keys[0]
    trial['resp'] = resp
    trial['rt'] = rt
    results.append(trial)
    
# win.close()
```

## 3. Data Logging

```python
df = pd.DataFrame(results)
df.to_csv('stroop_data.csv', index=False)
display(df)
```
