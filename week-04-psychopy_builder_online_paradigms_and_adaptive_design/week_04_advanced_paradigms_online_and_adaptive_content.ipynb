```python
from psychopy import data
import matplotlib.pyplot as plt
```

## 1. Staircase Logic
Using PsychoPy's StaircaseHandler for an Up-Down method.

```python
staircase = data.StaircaseHandler(startVal=0.5,
                                  stepType='lin',
                                  stepSizes=[0.1, 0.05],
                                  nTrials=20,
                                  nUp=1, nDown=3) # 3-down 1-up rule

# Simulation loop
print("Simulating responses...")
for contrast in staircase:
    # Simulate observer: correct if contrast > 0.3
    resp = 1 if contrast > 0.3 else 0
    staircase.addResponse(resp)
    
print(f"Final Threshold estimate: {staircase.intensities[-1]:.3f}")
```

## 2. Visualization

```python
plt.plot(staircase.intensities)
plt.xlabel('Trial')
plt.ylabel('Contrast')
plt.title('Ladder Plot')
plt.show()
```
