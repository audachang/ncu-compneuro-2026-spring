## 1. Data Loading
Using nilearn to fetch or load data.

```python
# Load a sample image if local file not present
# img = image.load_img('sub-01_task-rest_bold.nii.gz')

# Using Multi-Subject Dictionary Learning dataset as example
from nilearn import datasets
rest_dataset = datasets.fetch_adhd(n_subjects=1)
img = image.load_img(rest_dataset.func[0])
print(f"Image Shape (4D): {img.shape}")
```

## 2. Visualization

```python
plotting.plot_epi(image.mean_img(img), title="Mean EPI")
plotting.show()
```
